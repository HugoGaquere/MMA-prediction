## Count fights by fighter

```python
data = []

for _, fighter_details_row in fighter_details.iterrows():
    first_name = fighter_details_row['FIRST']
    last_name = fighter_details_row['LAST']

    # check that first_name and last_name types are string
    if not isinstance(first_name, str) or not isinstance(last_name, str):
        continue

    filtered_fights = fight_results[
        fight_results['BOUT'].str.contains(first_name) &
        fight_results['BOUT'].str.contains(last_name)
    ]


    if not filtered_fights.empty:
        wins, losses, no_contests, draws = 0, 0, 0, 0

        for _, fight_row in filtered_fights.iterrows():
            outcome = fight_row['OUTCOME']

            if outcome == 'NC/NC':
                no_contests += 1
            elif outcome == 'D/D':
                draws += 1
            else:
                bout = fight_row['BOUT']
                fighters = bout.split('vs.')
                outcome_position = 0 if first_name + ' ' + last_name in fighters[0] else 1
                outcome = outcome.split('/')[outcome_position]

                if outcome == 'W':
                    wins += 1
                elif outcome == 'L':
                    losses += 1

        nb_fights = len(filtered_fights)

        data.append([first_name + ' ' + last_name, nb_fights, wins, losses, no_contests, draws])
```

```python
# create new pandas dataframe containing 2 columns: "FIGHTER" and "NB FIGHTS"
df = pd.DataFrame(data, columns=['FIGHTER', 'NB FIGHTS', 'NB WINS', 'NB LOSSES', 'NB NC', 'NB DRAW'])
```
